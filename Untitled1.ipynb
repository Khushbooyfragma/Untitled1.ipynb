```python
# Design a simple calculator with basic arithmetic operation prompt the user to input two number and an operation choice perform the calculation and display the result
first = input ("enter first number : ")
# This line use to take first input number
operator = input ("enter operator (+,-,*,/,%) : ")
# This line use to take an operator user want to perform operation
second = input ("enter second number : ")
# This line use to take second input number
first = int(first)
second = int(second)
# Above line use to change string to integer
if operator == "+":
    print (first + second) #addition operation
elif operator == "-":
    print (first - second) #Substraction operation
elif operator == "*":
    print (first * second) #multiplication operation
elif operator == "/":
    print (first / second) #devide operation
elif operator == "%":
    print (first % second) #modulo operation
else:
    print("invalid operation")
```
